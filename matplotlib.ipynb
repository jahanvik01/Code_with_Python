```python
#Plotting x and y

import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(3, 3)) 
x = [1, 2, 3, 4]
y = [3, 7, 1, 5]

plt.plot(x, y)
plt.show()
```

```python
#Labels

plt.figure(figsize=(3, 3)) 
x = [1, 2, 3, 4]
y = [3, 7, 1, 5]

plt.plot(x, y)

plt.title("Plotting", loc='left')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
#Grid

plt.figure(figsize=(3, 3)) 
x = [1, 2, 3, 4]
y = [3, 7, 1, 5]

plt.plot(x, y)
plt.grid(axis='x', color='green', linestyle=':', linewidth=0.5)
plt.show()
```

```python
#Subplot, Markers, SuperTitle, Fontdict

font1={'family':'Times New Roman', 'color':'blue', 'fontsize':25}
font2={'family':'Times New Roman', 'color':'red', 'fontsize':15}

#plot 1:
x = [0, 1, 2, 3]
y = [3, 8, 1, 10]

plt.subplot(2, 1, 1)
plt.plot(x,y, marker='*', linestyle='dotted')
plt.title("Plot 1", fontdict=font2)

#plot 2:
x = [0, 1, 2, 3]
y = [10, 20, 30, 40]

plt.subplot(2, 1, 2)
plt.plot(x,y, marker='.', linestyle='dashed')
plt.title("Plot 2", fontdict=font2)

plt.suptitle("Super Title", fontdict=font1)
plt.show()
```

```python
import pandas as pd

data= pd.read_excel("medal count.xlsx")
df= pd.DataFrame(data)
#print(df.head(10))

countries = df['country_code']
gold = df['Gold Medal']
plt.subplot(2, 2, 1)  
plt.bar(countries, gold, color='hotpink')

countries = df['country_code']
gold = df['Gold Medal']
plt.subplot(2, 2, 2)
plt.plot(countries, gold, color='orange') 

countries = df['country_code']
gold = df['Gold Medal']
plt.subplot(2, 2, 3)
plt.scatter(countries, gold, color='orange') 

countries = df['country_code']
gold = df['Gold Medal']
plt.subplot(2, 2, 4)
plt.step(countries, gold, color='hotpink') 

plt.suptitle("Different Charts")
plt.xticks(rotation=30)
plt.show()
```

# aggregate functions

```python
# mean, mode, median

data= pd.read_csv("medals_total.csv")
df= pd.DataFrame(data)
#print(df.head(10))

gold = df['Gold Medal']
silver = df['Silver Medal']
bronze = df['Bronze Medal']

mean = np.mean(gold)
median = np.median(gold)
sum_value = np.sum(gold)

mean = np.mean(silver)
median = np.median(silver)
sum_value = np.sum(silver)

mean = np.mean(bronze)
median = np.median(bronze)
sum_value = np.sum(bronze)

plt.figure(figsize=(5, 3))   
plt.title('Data Visualization with Aggregates')
plt.plot(gold, label='Gold')
#plt.axhline(y=mean, color='r', linestyle='--', label='mean')
plt.axhline(y=median, color='g', linestyle='--', label='median')
plt.axhline(y=sum_value, color='m', linestyle='--', label='sum_value')
plt.legend()

plt.figure(figsize=(5, 3)) 
plt.plot(silver, label='Silver')
plt.axhline(y=mean, color='r', linestyle='--', label='mean')
#plt.axhline(y=median, color='g', linestyle='--', label='median')
plt.axhline(y=sum_value, color='m', linestyle='--', label='sum_value')
plt.legend()

plt.figure(figsize=(5, 3))   
plt.plot(bronze, label='Bronze')
plt.axhline(y=mean, color='r', linestyle='--', label='mean')
plt.axhline(y=median, color='g', linestyle='--', label='median')
#plt.axhline(y=sum_value, color='m', linestyle='--', label='sum_value')
plt.legend()

plt.show()
```

```python
# GroupBy

data= pd.read_excel("employee.xlsx")
df= pd.DataFrame(data)
# print(df)

gp = df.groupby("Department")["Bonus %"].sum()
# print(gp)

plt.figure(figsize=(5, 3))  
plt.bar(gp.index, gp.values, color='hotpink', alpha=0.7)
plt.title("Bonus% for different Departments", color= "m")
plt.xticks(rotation=30)
plt.show()
```
