# Hist plot

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = [23, 33, 45, 15, 67, 12 ,54]

plt.figure(figsize=(5, 3))
sns.histplot(df, kde=True)
plt.show()
```

```python
df = pd.read_csv('CoffeeChart.csv')

plt.figure(figsize=(5, 3))
sns.histplot(x="coffee_name", data=df, discrete="false", hue="cash_type", kde=True)
plt.xticks(rotation=45)
plt.show()
```

# Line plot

```python
x = [1,2,3,4,5]
y = [23,12,65,23,45]

df = pd.DataFrame({"Days":x, "No of people":y})

plt.figure(figsize=(5, 3))
sns.lineplot(x = "Days", y = "No of people", data = df)
plt.show()
```

```python
ds = pd.read_excel("employee.xlsx")

sns.lineplot(x="Department", y="Age", hue="Gender", data=ds)
plt.xticks(rotation=20)
plt.show()
```

```python
df = pd.read_csv('CoffeeChart.csv')

plt.figure(figsize=(5, 3))
sns.lineplot(x="coffee_name",y="money",data=df, hue="cash_type", palette="Set2_r")
plt.xticks(rotation=45)
plt.show()
```

# Lm plot

```python
df=sns.load_dataset("iris")

sns.lmplot(x="sepal_length", y="petal_length", data=df)
plt.show()
```

# Bar plot

```python
df = pd.read_excel('medal count.xlsx')

sns.barplot(x ='country_code', y ='Total', data = df)
plt.grid(alpha=0.2, color='red', linestyle='--')
plt.show()
```

```python
df = pd.read_csv('CoffeeChart.csv')

plt.figure(figsize=(5, 3))
sns.barplot(x="coffee_name", y="money", data =df, palette="bwr", hue="cash_type", )
plt.xticks(rotation=45)
plt.show()
```

# Count plot

```python
df=sns.load_dataset("tips")
df.head()
```

```python
plt.figure(figsize=(5, 3))
sns.countplot(x = "sex", data = df, palette ='plasma')
plt.show()
```

```python
plt.figure(figsize=(5, 3))
sns.countplot(x = "sex", data = df, hue="time", palette="spring", edgecolor="green", alpha=0.7)
plt.show()
```

# Box plot

```python
df = sns.load_dataset('tips')

plt.figure(figsize=(5, 3))
sns.boxplot(x ='day', y ='total_bill', data = df)
plt.show()
```

```python
plt.figure(figsize=(5, 3))
sns.boxplot(x="sex", y="tip", data=df, hue="smoker",palette="spring")
plt.show()
```

# Violoin plot

```python
plt.figure(figsize=(5, 3))
sns.violinplot(x=df["total_bill"], y=df["time"])
plt.show()
```

```python
sns.violinplot(x="sex", y="tip", data=df, hue='day', split=True)
plt.show()
```
