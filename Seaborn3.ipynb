# FacetGrid

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = sns.load_dataset('tips')

#scatter plot

mp = sns.FacetGrid(df, col='sex', hue='day')
mp.map(sns.scatterplot, 'total_bill', 'tip')
mp.add_legend()
plt.show()
```

```python
#Bar plot

mp = sns.FacetGrid(df, col='day')
mp.map(sns.barplot, 'sex', 'tip', order=['Male', 'Female'], color= 'pink')
plt.show()
```

```python
#Hist plot

mp = sns.FacetGrid(df, row ='smoker', col ='time')  
mp.map(plt.hist, 'total_bill', bins = 10, color='grey', alpha=0.8, edgecolor='blue') 
plt.show() 
```

```python
#Regression plot

mp = sns.FacetGrid(df, col ='time', hue ='smoker') 
mp.map(sns.regplot, "total_bill", "tip") 
mp.add_legend()
plt.show()
```

```python
mp = sns.PairGrid(df, hue='time') 
mp.map_diag(sns.histplot)
mp.map_offdiag(sns.scatterplot)
plt.show()
```

```python
mp = sns.FacetGrid(df, col ='time') 
mp.map(sns.kdeplot, 'tip') 
plt.show()
```

# Kernel Density Estimation (KDE) plot

```python
df = sns.load_dataset('tips')
sns.kdeplot(y='total_bill', data = df)
plt.show()
```

```python
sns.kdeplot(x='size', hue='day', data = df)
plt.show()
```

```python
sns.kdeplot(x='tip', hue='day', fill=True, data = df)
plt.show()
```

```python
sns.kdeplot(x='total_bill', y='tip', data = df)
plt.show()
```

```python
sns.kdeplot(x='total_bill', y='tip', fill=True, cmap='GnBu', data = df)
plt.show()
```
