# Stem Plot

```python
import matplotlib.pyplot as plt

x= [2, 5, 7, 3, 1, 4, 6, 5]
y= [10, 40, 60, 25, 40, 30, 70, 10]

plt.figure(figsize=(3, 3))
plt.stem(x, y)
plt.show()
```

```python
x= [2, 5, 7, 3, 1, 4, 6, 5]
y= [10, 40, 60, 25, 40, 30, 70, 10]

plt.figure(figsize=(3, 3))
plt.stem(x, y, basefmt = 'green', orientation ="horizontal", linefmt=":", markerfmt="D")
plt.show()
```

```python
import pandas as pd

data = pd.read_excel("employee.xlsx")
df = pd.DataFrame(data)
df1= df.head(50)


plt.stem(df1["Age"])
plt.show()
```

```python
data = pd.read_excel("employee.xlsx")
df = pd.DataFrame(data)
df1= df.head(20)

plt.figure(figsize=(6, 3))

plt.plot(df1["Age"])
plt.stem(df1["Age"], linefmt="--", markerfmt="r*")
plt.title("Stem Plot with Line Plot", loc='left')
plt.show()
```

# Box Plot

```python
data = [10,12,23,28,32,44]

plt.figure(figsize=(3, 3))
plt.boxplot(data)
plt.show()
```

```python
import numpy as np

d1= [23, 44, 12, 32, 10, 28, 10, 8, 25]
d2= [2, 4, 6, 8, 10, 15, 30]
d3=np.sort(d1)

bx = (d3, d2)
plt.figure(figsize=(4, 4))
plt.boxplot(bx, showmeans=True, capprops=dict(color='blue'), whiskerprops=dict(color='b'))
plt.show()
```

```python
data= pd.read_excel("car_data.xlsx")
df= pd.DataFrame(data)
# print(df)

plt.figure(figsize=(3, 3))
plt.boxplot(df["Engine (CC)"], labels=["Engine"])
plt.show()
```

# Stack Plot

```python
data= pd.read_csv("medals_total.csv")
df= pd.DataFrame(data)
#print(df.head(10))

countries = df['country_code']
gold = df['Gold Medal']
silver = df['Silver Medal']
bronze = df['Bronze Medal']

l = ['Gold', 'Silver', 'Bronze']

plt.figure(figsize=(5, 4))
plt.title("Count of Medals")
plt.stackplot(countries, gold, silver, bronze, labels=l, colors=['yellow','grey', 'orange'])
plt.legend()
plt.show()
```

# Step Plot

```python
plt.figure(figsize=(4, 4))
plt.title("Count of Silver Medals")
plt.step(countries, silver, color='m', marker='o')
plt.show()
```

# Fill_Between Plot

```python
x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]

plt.figure(figsize=(4, 4))
plt.plot(x, y, color="green")
plt.fill_between(x=[2,4], y1=2, y2=4, color="lightgreen")

plt.title("fill_between plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```
