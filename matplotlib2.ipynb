# Scatter Plot

```python
import matplotlib.pyplot as plt

x = [1,3,5,2,7,6]
y = [3,1,6,8,3,2]

plt.figure(figsize=(3, 3))
plt.scatter(x,y)

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x = [1,3,5,2,7,6]
y = [3,1,6,8,3,2]
colors = (["green","yellow","orange","purple","brown","magenta"])
sizes = [200,350,80,50,100,290]

plt.figure(figsize=(3, 3))
plt.scatter(x,y, c=colors, s=sizes, alpha=0.5)

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
import numpy as np
import pandas as pd

color=np.random.randint(10, 100, 1000)
data = pd.read_excel("employee.xlsx")

df= pd.DataFrame(data)

plt.figure(figsize=(7, 3))
plt.scatter(df['Age'], df['Bonus %'], c=color, cmap='hot')
plt.show()
```

# Bar Plot

```python
import matplotlib.pyplot as plt

x = ["Python", "C", "C++", "JAVA"]
y = [85, 70, 60, 77]

plt.figure(figsize=(3, 3)) 
plt.xlabel("language", fontsize=12)
plt.ylabel("popularity", fontsize=12)

plt.barh(x, y)   #for horizontal bar
plt.show()
```

```python
x = ["Python", "C", "C++", "JAVA"]
y = [85, 70, 60, 77]

plt.figure(figsize=(3, 3)) 
plt.xlabel("language", fontsize=12)
plt.ylabel("popularity", fontsize=12)

plt.bar(x, y, color='lightgreen', width = 0.3, align="edge", edgecolor="b", linewidth=2, label="my_graph")
plt.legend()
plt.grid(axis='y', alpha=0.3, color='r', linestyle='--', linewidth=0.5)
plt.show()
```

```python
import numpy as np

x = ["Python", "C", "C++", "JAVA"]
y = [85, 70, 60, 77]
z = [20, 45, 51, 37]

width=0.2
p = np.arange(len(x))
p1 = [j + width for j in p]

plt.figure(figsize=(3, 3)) 
plt.xlabel("Language")
plt.ylabel("Numbers")

plt.bar(p,y,width,color="hotpink", align='edge', edgecolor='orange')
plt.bar(p1,z,width,color="orange", edgecolor='hotpink')

plt.xticks(p+width/2, x, rotation=20)
plt.show()
```

# Histogram

```python
import random 

x = np.random.randint(0,60,(50))

plt.figure(figsize=(3, 3)) 
plt.hist(x)
plt.show()
```

```python
x = np.random.randint(0,60,(50))

plt.figure(figsize=(3, 3)) 
plt.hist(x, color="lightgreen", alpha=0.8, edgecolor="black")
plt.show()
```

```python
x = np.random.randint(0,60,(50))
l = [5,10,15,20,25,30]

plt.figure(figsize=(3, 3)) 
plt.hist(x, color="lightgreen", edgecolor="black", bins=l)
plt.grid(axis='x', alpha=0.3, color='b', linestyle='--', linewidth=0.5)
plt.show()
```

# Pie Chart

```python
cars = ['AUDI', 'BMW', 'FORD', 'TESLA', 'JAGUAR', 'MERCEDES']
data = [23, 17, 35, 29, 12, 41]
c = ['r', 'g', 'b', 'c', 'm', 'y']

plt.pie(data, labels=cars, colors = c, radius=0.8)
plt.show()
```

```python
cars = ['AUDI', 'BMW', 'FORD', 'TESLA', 'JAGUAR', 'MERCEDES']
data = [23, 17, 35, 29, 12, 41]
myexplode = [0.2, 0, 0, 0, 0, 0]

plt.pie(data, labels=cars, explode=myexplode, shadow= True, autopct="%.1f%%")
#plt.legend(title = "Cars:")
plt.show()
```

```python
cars = ['AUDI', 'BMW', 'FORD', 'TESLA', 'JAGUAR', 'MERCEDES']
data = [23, 17, 35, 29, 12, 41]
data2 = [45, 13, 24, 67, 33, 19]
myexplode = [0, 0, 0, 0, 0, 0]

plt.pie(data, labels=cars, radius=1.5)
plt.pie(data2, radius=1, explode=myexplode, wedgeprops={'linewidth':2, 'edgecolor':'w'})
plt.show()
```
